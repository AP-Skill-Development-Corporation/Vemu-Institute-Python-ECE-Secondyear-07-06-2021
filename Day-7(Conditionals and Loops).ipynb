```python
z=1
while z<=20: # true
    print(z,end=" ")
```

```python
z=1
while z<=20: # true   
    print(z,end=" ")
    z=z+1
    
    
    # 1  2 3 4.........19 20 21 
```

```python
##  sum of digits in while loop

#12345: 1+2+3+4+5=15

# 12345/10 =1234  ,rem =5
# 1234/10 = 123   ,rem =4
# 123/10 = 12   ,rem=3
# 12/10  =1  , rem=2
# 1/10=rem =1



x=int(input("enter  any number"))
y=0
while x!=0:
    rem=x%10
    y+=rem # y=y+rem
    x//=10 
    
print("Sum of the digits"  , y)
    
```
